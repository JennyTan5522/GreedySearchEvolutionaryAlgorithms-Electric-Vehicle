```python
import numpy as np
from scipy.spatial import distance
from scipy import spatial
import random 
```

```python
#This one maybe can change to RANDOM read if got time !! (Try not to sequential read -> reduce time)
'''Read the problem instance and generate the initial object vector'''
with open('evrp-benchmark-set/E-n22-k4.evrp','r') as f:
    data=f.read().splitlines()  
    
#Store node and demand as a dictionary for better access
NODE={}
DEMAND={}
for idx,line in enumerate(data):
    record=line.split(':')
    record[0]=record[0].strip()

    if (record[0]=='OPTIMAL_VALUE'):
        OPTIMUM=float(record[1].strip())

    if (record[0]=='VEHICLES'):
        MIN_VEHICLES=int(record[1].strip())

    if (record[0]=='DIMENSION'):
        PROBLEM_SIZE=int(record[1].strip()) 

    if (record[0]=='STATIONS'):
        NUM_OF_STATIONS=int(record[1].strip())

    if (record[0]=='CAPACITY'):
        MAX_CAPACITY=int(record[1].strip())

    if (record[0]=='ENERGY_CAPACITY'):
        BATTERY_CAPACITY=int(record[1].strip())

    if (record[0]=='ENERGY_CONSUMPTION'):
        ENERGY_CONSUMPTION=float(record[1].strip())

    if (record[0]=='NODE_COORD_SECTION'):
        NUM_OF_CUSTOMERS=PROBLEM_SIZE-1
        ACTUAL_PROBLEM_SIZE=PROBLEM_SIZE+NUM_OF_STATIONS
        idx+=1
        for i in range(ACTUAL_PROBLEM_SIZE):
            node_data=data[idx+i]
            node_data=node_data.split(' ')
            #Save node as tuple (index,x,y)
            NODE[int(node_data[0])]=(int(node_data[1]),int(node_data[2]))

    if (record[0]=='DEMAND_SECTION'):
        idx+=1
        for i in range(PROBLEM_SIZE):
            DEMAND[int(data[idx+i].split(' ')[0])]=int(data[idx+i].split(' ')[1])
```

```python
def euclidean_distance(node1:int,node2:int): 
    '''Compute and return the euclidean distance of 2 coordinates'''
    coor1=(NODE[node1][0],NODE[node1][1]) #(x,y)
    coor2=(NODE[node2][0],NODE[node2][1])
    return distance.euclidean(coor1,coor2)
#euclidean_distance(5,7)
```

```python
def compute_distances(matrix):
    '''Compute the distance matrix of the problem instance'''
    for i in range(ACTUAL_PROBLEM_SIZE): #Eg:Loop from index 0(node1) to index 29(node30)
        for j in range(ACTUAL_PROBLEM_SIZE):
            matrix[i][j]=euclidean_distance(i+1,j+1)
    return matrix
```

```python
def generate_2D_distance_matrix(NODE:int,ACTUAL_PROBLEM_SIZE:int):
    '''Generate 2D distance matrix and find the distance between 2 points'''
    #Initialize 2D array
    matrix=np.zeros((len(NODE),len(NODE)))

    #Calculate euclidean distance based on 2 points
    distanceMatrix=compute_distances(matrix)
    
    return distanceMatrix
```

```python
distanceMatrix=generate_2D_distance_matrix(NODE,ACTUAL_PROBLEM_SIZE)
```

```python
#unclusterNodeList=List of customer node (index) -> Empty then stop iter (If assign one customer to that part cluster then remove cust from list)
#finalCluster=2D(more easier) to store cluster and cluster members, first cluster 5,8,2,4; second cluster 1,3,7
#Before add to cluster check capacity

#1. seedpoint, then remove seedpoint from list because it already belongs to one cluster
#2. currentCluster=New seedpoint then New list to keep track (cluster members+seedpoint)
#3. currentCapacity (starting is seedpoint capacity) exclude depot
#3. While len(unclusterNodeList)>0 (Exclude starting point) 
#4.      Find nearest neigbour of the seedpoint (When sorting then exclude those not in uncluster node one->Means already got cluster(must make sure the node if not in cluster first))
#5.           If (check currentCapacity<maxCapacity) then only update currentCapacity, currentCluster(nearest neigbour), remove node from unclusterNodeList
#6.        else then find a new seedPoint (Remove seedPoint from unclusterList) -> Take currentCluster save into finalCluster, refresh currentCluster(start with newSeedPoint) and currentCapacity
```

```python
#List used to keep track uncluster customer node (Exclude depot+charging stations) eg:2-22
unclusterNodeList=list(NODE.keys())[1:NUM_OF_CUSTOMERS+1] 
print(f'unclusterNodeList before remove seedPoint: {unclusterNodeList}')

#2D list to store each cluster members
finalCluster=[] 

#1.Randomly select a customer as seedPoint in a cluster (Customer node range 2-22)
seedPoint=random.randint(2,NUM_OF_CUSTOMERS+1) 

#Remove seedPoint from unclusterNodeList
unclusterNodeList.remove(seedPoint)

#currentCluster: Keep track the current cluster's members based on the seedPoint given (seedPoint+cluster members)
currentCluster=[seedPoint] 

#currentCapacityDemand: Keep track of the current capacity demands of the given customer node 
currentCapacityDemand=DEMAND[seedPoint]

print('===========================================================')
print(f'seedPoint             : {seedPoint}')
print(f'unclusterNodeList     : {unclusterNodeList}')
print(f'currentCluster        : {currentCluster}')
print(f'currentCapacityDemand : {currentCapacityDemand}')
print(f'depot + customers     : {len(distanceMatrix[seedPoint-1][:NUM_OF_CUSTOMERS+1])}')
print('===========================================================')

#Extract distance of depot+customers
sortCustIdxMatrix=np.argsort(distanceMatrix[seedPoint-1][:NUM_OF_CUSTOMERS+1]) 
print(f'sortCustIdxMatrix : {sortCustIdxMatrix}')

#Node used to keep track customerNode
#Exclude seedPoint itself (distance=0) and depot, we can straight jump to second record
availableCustNode=[sortCustIdxMatrix[i]+1 for i in range(1,sortCustIdxMatrix.shape[0]) if sortCustIdxMatrix[i]!=0] 
            
print(f'availableCustNode : {availableCustNode}')
idx=0
while(len(availableCustNode)>0):
    #2.2-Check whether if the currentCapacity < maxCapacity, if yes then only update currentCapacity + remove node from unclusterNodeList
    cust=availableCustNode[idx]
    print(cust,idx)
    print(f'Customer {cust}, {DEMAND[cust]}')
    if(currentCapacityDemand+DEMAND[cust] < MAX_CAPACITY):
        #Update currentCapacity
        currentCapacityDemand+=DEMAND[cust]  
            
        #Add node into currentCluster
        currentCluster.append(cust)
            
        #Remove members from availableNode
        availableCustNode.remove(cust)
            
        print(f'CurrentCapacityDemand : {currentCapacityDemand}')
        print(f'CurrentCluster        : {currentCluster}')
        print(f'AvailableCustNode     : {availableCustNode}')
        print('------------------------------------------------------')
    else:
        #Add currentCluster to finalCluster
        finalCluster.append(currentCluster)
        print('***************************************************************')
        print(f'FinalCluster          : {finalCluster}')

        #Create a new cluster based on new seedPoint
        while(True):
            seedPoint=random.randint(2,NUM_OF_CUSTOMERS+1)
            print(f'newSeedPoint          : {seedPoint}')
            print(f'availableCustNode     : {availableCustNode}')
                
            #Make sure the seedPoint is in the availableCustNode
            if (seedPoint in availableCustNode):
                availableCustNode.remove(seedPoint)
                
                #Create new currentCluster
                currentCluster=[seedPoint]
                    
                #Update currentCapacityDemand based on the newSeedPoint
                currentCapacityDemand=DEMAND[seedPoint]
                print('\n            == NEW ==')
                print(f'currentCluster        : {currentCluster}')
                print(f'currentCapacityDemand : {currentCapacityDemand}')
                print(f'availableCustNode     : {availableCustNode}')
                break
                
#Update the last currentCluster to finalCluster
finalCluster.append(currentCluster) 
```
